```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\TCC\\ScriptsPypsaBrazil\\open-brazilian-energy-data_atualizado\\hydro_inflow\\raw\\raw_2000_2020_daily_N_Simples_Energia_Natural_Afluente_Subsistema_Ano_Fu_data.csv",
                encoding='utf-16-le',
                sep='\t')

df = df.T
df.reset_index(drop=True, inplace=True)
df.index = range(len(df))
df.index.name = 'index'
df = df.drop(index=[0, 1, 2]).reset_index(drop=True)
df["Subsistema"] = df.iloc[:,1].iloc[0]
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.rename(columns={df.columns[0]: 'Din Instante', df.columns[1]: 'Val Enaarmazenavelmwmes'}, inplace=True)
df['index'] = range(len(df))
reorder = ['index', 'Subsistema', 'Din Instante', 'Val Enaarmazenavelmwmes']
df = df[reorder].copy()
df['Val Enaarmazenavelmwmes'] = df['Val Enaarmazenavelmwmes'].str.replace('.', '').astype(float)
portuguese_month_abbreviations = {
    'jan.': 'jan',
    'fev.': 'feb',
    'mar.': 'mar',
    'abr.': 'apr',
    'mai.': 'may',
    'jun.': 'jun',
    'jul.': 'jul',
    'ago.': 'aug',
    'set.': 'sep',
    'out.': 'oct',
    'nov.': 'nov',
    'dez.': 'dec'
}

# Convert the 'Date_Column' to datetime format
df['Din Instante'] = pd.to_datetime(
    df['Din Instante'].replace(portuguese_month_abbreviations, regex=True),
    format='%d de %b de %y',
)

# Format the 'Date_Column' in the desired format
df['Din Instante'] = df['Din Instante'].dt.strftime('%d.%m.%Y %H.%M.%S')

df.head()
```
