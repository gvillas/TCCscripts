```python
df = pd.read_csv("C:\\TCC\\ScriptsPypsaBrazil\\open-brazilian-energy-data_atualizado\\load_time_series\\Simples Carga de Energia Dia Hora - Norte.csv",
                encoding='utf-16-le',
                sep='\t')
df = df.T
df.reset_index(drop=True, inplace=True)
df.index = range(len(df))
df.index.name = 'index'
df["Subsistema"] = df.iloc[:,1].iloc[0]
df.drop(0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.rename(columns={df.columns[0]: 'Din Instante', df.columns[1]: 'Val Cargaenergiaconmwmed'}, inplace=True)
df['index'] = range(len(df))
reorder = ['index', 'Subsistema', 'Din Instante', 'Val Cargaenergiaconmwmed']
df = df[reorder].copy()
df['Val Cargaenergiaconmwmed'] = df['Val Cargaenergiaconmwmed'].str.replace('.', '').astype(float)
portuguese_month_abbreviations = {
    'jan': 'jan',
    'fev': 'feb',
    'mar': 'mar',
    'abr': 'apr',
    'mai': 'may',
    'jun': 'jun',
    'jul': 'jul',
    'ago': 'aug',
    'set': 'sep',
    'out': 'oct',
    'nov': 'nov',
    'dez': 'dec'
}

# Convert the 'Date_Column' to datetime format
df['Din Instante'] = pd.to_datetime(
    df['Din Instante'].replace(portuguese_month_abbreviations, regex=True),
    format='%d  de %b.  de %y - %H h',
)

# Format the 'Date_Column' in the desired format
df['Din Instante'] = df['Din Instante'].dt.strftime('%d.%m.%Y %H.%M.%S')


df.head()
```
